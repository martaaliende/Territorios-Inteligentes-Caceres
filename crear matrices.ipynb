### matriz distancias km

```python
# 2. Geolocalizar los municipios
geo_output = gpd.tools.geocode(
    [f"{m}, Cáceres, Extremadura, Spain" for m in municipios]
)
coords = {}
for i, row in geo_output.iterrows():
    nombre = municipios[i]
    coords[nombre] = (row.geometry.y, row.geometry.x)  # (Lat, Lon)

# 3. Descargar el grafo de la Provincia de Cáceres
print("Descargando el grafo de carreteras de Cáceres...")
G = ox.graph_from_place('Provincia de Cáceres, Spain', network_type='drive')
G = ox.add_edge_speeds(G)
G = ox.add_edge_travel_times(G)

# 4. Encontrar el nodo más cercano para cada municipio
nodos_municipios = {}
for nombre, (lat, lon) in coords.items():
    nodos_municipios[nombre] = ox.distance.nearest_nodes(G, lon, lat)

# 5. Calcular la matriz de distancias
nombres = list(nodos_municipios.keys())
matriz_distancias = pd.DataFrame(0.0, index=nombres, columns=nombres)

print("Calculando rutas...")
for i, origen in enumerate(nombres):
    print(f"  Procesando {i+1}/{len(nombres)}: {origen}")
    nodo_origen = nodos_municipios[origen]

    # Una sola pasada de Dijkstra desde este origen a todos los nodos
    lengths = nx.single_source_dijkstra_path_length(G, nodo_origen, weight='length')

    for destino in nombres:
        if origen != destino:
            nodo_destino = nodos_municipios[destino]
            dist_m = lengths.get(nodo_destino, None)
            if dist_m is not None:
                matriz_distancias.loc[origen, destino] = round(dist_m / 1000, 2)
            else:
                matriz_distancias.loc[origen, destino] = None
# matriz_distancias creada
```

### matriz tiempo minutos

```python
# 2. Geolocalizar los municipios
print("Geolocalizando municipios...")
# geo_output = gpd.tools.geocode(municipios)
geo_output = gpd.tools.geocode(
    [f"{m}, Cáceres, Extremadura, Spain" for m in municipios]
)
# Creamos un diccionario con las coordenadas (Latitud, Longitud)
# Geopandas devuelve geometry como (Long, Lat), así que extraemos .y (Lat) y .x (Long)
coords = {}
for i, row in geo_output.iterrows():
    nombre = municipios[i]
    coords[nombre] = (row.geometry.y, row.geometry.x) # (Lat, Lon)

# 3. Descargar el grafo de la Provincia de Cáceres
# Usamos 'drive' para rutas en coche.
print("Descargando el grafo de carreteras de Cáceres (esto puede tardar unos segundos)...")
G = ox.graph_from_place('Provincia de Cáceres, Spain', network_type='drive')

# Añadir velocidades y tiempos a las aristas del grafo
G = ox.add_edge_speeds(G)
G = ox.add_edge_travel_times(G)

# 4. Encontrar el nodo más cercano en el grafo para cada municipio
# (El grafo no conoce las direcciones, solo puntos en la red, por eso buscamos el nodo más cercano)
nodos_municipios = {}
for nombre, (lat, lon) in coords.items():
    # nearest_nodes espera (Long, Lat) -> (x, y)
    nodo_cercano = ox.distance.nearest_nodes(G, lon, lat)
    nodos_municipios[nombre] = nodo_cercano

# 5. Calcular las matrices de distancia y tiempo
nombres = list(nodos_municipios.keys())
num_municipios = len(nombres)

# Inicializamos los DataFrames con ceros
matriz_tiempos = pd.DataFrame(0.0, index=nombres, columns=nombres)

print("Calculando rutas...")
for i in range(num_municipios):
    for j in range(num_municipios):
        origen = nombres[i]
        destino = nombres[j]
        nodo_origen = nodos_municipios[origen]
        nodo_destino = nodos_municipios[destino]

        if i != j:
            try:
                # Calcular tiempo de la ruta más rápida (en segundos)
                time_s = nx.shortest_path_length(G, nodo_origen, nodo_destino, weight='travel_time')

                # Guardar en las matrices convirtiendo unidades
                matriz_tiempos.loc[origen, destino] = round(time_s / 60, 1)      # a Minutos
            except nx.NetworkXNoPath:
                # En caso de que no haya conexión (raro en una misma provincia)
                matriz_tiempos.loc[origen, destino] = None

# matriz_tiempos creada
```

## Crear matriz de adyacencia

```python
import geopandas as gpd
import pandas as pd

# 1. CARGAR SHAPEFILE
# https://centrodedescargas.cnig.es/CentroDescargas/limites-municipales-provinciales-autonomicos

ruta_shp = '/content/drive/MyDrive/Colab Notebooks/TFG/SHP_ETRS89/recintos_municipales_inspire_peninbal_etrs89'
gdf = gpd.read_file(ruta_shp)

# 2. CORREGIR NOMBRES
correcciones = {
    "Aldehuela del Jerte": "Aldehuela de Jerte"
}
gdf["NAMEUNIT"] = gdf["NAMEUNIT"].replace(correcciones)

# 3. FILTRAR MUNICIPIOS DE INTERÉS
gdf = gdf[gdf["NAMEUNIT"].isin(municipios)].copy()

# 4. ASEGURAR ORDEN
gdf = gdf.set_index("NAMEUNIT").loc[municipios].reset_index()

# 5. LIMPIEZA GEOMÉTRICA
gdf["geometry"] = gdf["geometry"].buffer(0)


# 6. MATRIZ DE ADYACENCIA
n = len(gdf)
adj_matrix = pd.DataFrame(0, index=municipios, columns=municipios)

# 7. CÁLCULO DE ADYACENCIA (FRONTERA REAL)
for i in range(n):
    geom_i = gdf.geometry.iloc[i]
    name_i = gdf.NAMEUNIT.iloc[i]

    for j in range(i + 1, n):
        geom_j = gdf.geometry.iloc[j]
        name_j = gdf.NAMEUNIT.iloc[j]

        # SOLO vecinos reales (comparten frontera)
        if geom_i.touches(geom_j):
            adj_matrix.loc[name_i, name_j] = 1
            adj_matrix.loc[name_j, name_i] = 1
# 8. DIAGONAL A 0
for m in municipios:
    adj_matrix.loc[m, m] = 0

# adj_matrix creada
```
